## 6. NLTK Data

```python
# Test NLTK data packages
try:
    import nltk.data
    
    # Check for punkt tokenizer
    nltk.data.find('tokenizers/punkt')
    print("✅ NLTK punkt tokenizer available")
    
    # Check for stopwords
    from nltk.corpus import stopwords
    english_stops = set(stopwords.words('english'))
    assert 'the' in english_stops
    print("✅ NLTK stopwords available")
    
    # Test tokenization
    from nltk.tokenize import word_tokenize, sent_tokenize
    text = "Hello world! This is a test sentence."
    words = word_tokenize(text)
    sentences = sent_tokenize(text)
    
    assert len(words) > 0
    assert len(sentences) > 0
    print(f"   Sample tokenization: {words}")
    print("✅ NLTK tokenization test passed")
    
except LookupError as e:
    print(f"❌ NLTK data missing: {e}")
    print("Run the following commands:")
    print("  python -c \"import nltk; nltk.download('punkt'); nltk.download('stopwords')\"")
```

## 9. Complete NLP Pipeline Test

```python
# Comprehensive NLP pipeline test
sample_text = """
Natural Language Processing (NLP) is a fascinating field that combines computer science, 
artificial intelligence, and linguistics. Companies like Google, Apple, and Microsoft 
have invested billions in NLP research. The University of Maryland offers excellent 
courses in this domain. Students often find NLP challenging but rewarding!
"""

print("🔍 Running complete NLP pipeline test...\n")

# 1. Basic text statistics
print("1. TEXT STATISTICS")
print(f"   Character count: {len(sample_text)}")
print(f"   Word count: {len(sample_text.split())}")
print(f"   Reading level: {flesch_reading_ease(sample_text):.1f}")

# 2. Language detection
print("\n2. LANGUAGE DETECTION")
detected_lang = detect(sample_text)
print(f"   Detected language: {detected_lang}")

# 3. spaCy processing
if 'nlp_sm' in locals():
    print("\n3. SPACY PROCESSING")
    doc = nlp_sm(sample_text)
    
    # Tokens and POS tags
    tokens_pos = [(token.text, token.pos_) for token in doc if not token.is_space]
    print(f"   First 10 tokens with POS: {tokens_pos[:10]}")
    
    # Named entities
    entities = [(ent.text.strip(), ent.label_) for ent in doc.ents]
    print(f"   Named entities: {entities}")
    
    # Dependency relations (first sentence only)
    first_sent = list(doc.sents)[0]
    dependencies = [(token.text, token.dep_, token.head.text) 
                   for token in first_sent if not token.is_space]
    print(f"   Dependencies (first sentence): {dependencies[:5]}...")  # Show first 5

# 4. NLTK processing
print("\n4. NLTK PROCESSING")
try:
    from nltk.tokenize import word_tokenize, sent_tokenize
    from nltk.corpus import stopwords
    
    # Sentence tokenization
    sentences = sent_tokenize(sample_text)
    print(f"   Number of sentences: {len(sentences)}")
    
    # Word tokenization and stopword removal
    words = word_tokenize(sample_text.lower())
    stop_words = set(stopwords.words('english'))
    filtered_words = [w for w in words if w.isalnum() and w not in stop_words]
    
    # Word frequency
    from collections import Counter
    word_freq = Counter(filtered_words)
    print(f"   Most common words: {word_freq.most_common(5)}")
    
except Exception as e:
    print(f"   NLTK processing failed: {e}")

print("\n🎉 SETUP TEST COMPLETED SUCCESSFULLY!")
print("\nYour environment is ready for NLP work. You can now:")
print("- Process text with spaCy and NLTK")
print("- Use machine learning libraries")
print("- Work with transformer models")
print("- Create visualizations")
print("\nIf you see this message, take a screenshot for your assignment submission!")
```

# Context-Free Grammars

The most commonly used mathematical system for modeling constituent structure in English and other natural languages is the **Context-Free Grammar**, or **CFG**. Context-Free Grammars are also called **Phrase-Structure Grammars**.

# Formal Definition of Context-Free Grammar

**Context-Free (Phrase Structure) Grammars** are a rewriting system that form a generative model (4-tuple):

- $\Sigma$: set of **terminal symbols** that are observable (disjoint from N)
- $N$: set of **non-terminal symbols** (or variables)
- $R$: set of **rules of production** (context-free rewriting rules):  
  each rule have the form $A \rightarrow \beta$; s.t. $A$ is a non-terminal, $\beta$ is a string of symbols from the finite set of string $(\Sigma \cup N)^*$
- $S$: designated **start symbol**

# Agreement

**Not ok**: The students (PL) loves (SG) the book

# Combinatory Categorial Grammar Basics

**"Slash category"** is essentially a nonterminal that is missing something to be complete, e.g. VP/NP (VP missing an NP to its right). Combinatory Categorial Grammar uses a similar idea, without context-free rewrite rules.

Intuitively, a categorial grammar operates by having a limited set of atomic categories (NP, S, PP, …) but then allowing lexical items to have more complex categories – one 1-argument functions

**Eg.**
$$\text{likes} := (S\backslash NP) / NP$$

A transitive verb is a function from NPs into predicate S. That is, it accepts two NPs as arguments and result in S.

```
    Mary      likes     musicals
     NP    (S\NP)/NP       NP
```

**Function application**

Forward app:  >  :  X / Y  Y → X  
(S\NP) / NP  NP → NP

Backward app: <  :  Y  X \ Y → X  
NP  S\NP → S

# CKY: Dynamic Programming Again!

**Dynamic programming**: split the problem to sub-problems and solve them, when complete, the sub-problem solutions solve the problem as whole.

**CKY** is applicable to CFGs that are in **Chomsky Normal Form**

All rules have the form $X \rightarrow x$ (preterminal to terminal) or $X \rightarrow YZ$

## Illustration of conversion to Chomsky Normal Form

Situations we need to address to convert to Chomsky Normal Form:

**Rules that mix non-terminal and terminal**  
INF-VP → to VP ~> INF-VP → TO VP

**Rules with a single non terminal (unite production)**  
A → B & B → y (non unite production) ~> A → y

**Rules with right hand-side longer than 2**  
A → BCy ~> X1 → BC & A→ X1y

# CKY

**Recognition**: Is there an S in cell [0, n]?

**Parsing**: Record all ways to arrive at S in [0, n] given the input

CKY was first O(n³) algorithm to parse CFGs, but limited to CNF

Earley then introduced the first O(n³) algorithm for arbitrary CFGs

# Probabilistic CKY

What about the goal of finding the best parse, if there's ambiguity?

Use **probabilistic CFG (PCFG)**, which is a CFG where rules are augmented with probabilities

**Model parameters**: such that

Probability of this NP spanning from i to j is the product of three probabilities:

- **P1** (probability of having a Det spanning from i to k)
- **P2** (probability of having an N spanning from k to j)
- **P3** (probability of rule NP → Det N given that we are expanding NP)

```
      NP
   Det   N
   the   dog
  (P1)  (P2)
     (P3)
i    k    j
```

Probability of any NP spanning from i to j is the sum of such probabilities, for all the different ways you could build an NP spanning from i to j

**Viterbi CKY parsing** is directly analogous to Viterbi decoding for HMMs: if you want the most probable latent structure, simply replace sum with max in the dynamic programming algorithm.

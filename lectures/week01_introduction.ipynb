## 2. What is Natural Language Processing?

> **Natural Language Processing (NLP)**: A field of computer science and artificial intelligence concerned with getting computers to understand, interpret, and generate human language in a valuable way.

### Core Challenges

1. **Ambiguity**
   - Lexical: "bank" (financial institution vs. river bank)
   - Syntactic: "I saw the man with the telescope"
   - Semantic: "The chicken is ready to eat"

2. **Context Dependency**
   - "It's cold in here" (statement vs. request to close window)
   - Pronouns, discourse, pragmatics

3. **Variability**
   - Different languages, dialects, styles
   - Informal vs. formal language
   - Evolution of language over time

4. **Creativity**
   - Novel word combinations
   - Metaphor, humor, sarcasm
   - Cultural references

## 3. The NLP Pipeline

### Traditional NLP Pipeline

```
Raw Text → Tokenization → Part-of-Speech Tagging → Parsing → Semantic Analysis → Application
```

### Core Components

1. **Tokenization**
   - Split text into tokens (words, subwords, characters)
   - Handle punctuation, contractions, special cases

2. **Morphological Analysis**
   - Analyze word structure (stems, prefixes, suffixes)
   - Lemmatization vs. stemming

3. **Part-of-Speech (POS) Tagging**
   - Assign grammatical categories (noun, verb, adjective, etc.)
   - Context-dependent classification

4. **Syntactic Analysis (Parsing)**
   - Identify grammatical structure
   - Constituency vs. dependency parsing

5. **Semantic Analysis**
   - Extract meaning from text
   - Named entity recognition, relation extraction

6. **Discourse Analysis**
   - Understand text structure beyond sentences
   - Coreference resolution, coherence

````python
# Demo: Simple NLP Pipeline with spaCy
import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Apple Inc. was founded by Steve Jobs in Cupertino, California."

# Process the text
doc = nlp(text)

print("Tokenization and POS Tagging:")
for token in doc:
    print(f"{token.text:<12} {token.pos_:<8} {token.lemma_:<12} {token.is_stop}")

print("\nNamed Entities:")
for ent in doc.ents:
    print(f"{ent.text:<15} {ent.label_:<10} {ent.start:<3} {ent.end}")

print("\nDependency Relations:")
for token in doc:
    print(f"{token.text:<12} --{token.dep_}--> {token.head.text}")
````

### Modern Approach: End-to-End Learning

**Traditional**: Modular pipeline with hand-crafted features
- Pro: Interpretable, modular, debuggable
- Con: Error propagation, feature engineering bottleneck

**Modern**: Neural end-to-end models
- Pro: Automatic feature learning, superior performance
- Con: Less interpretable, requires more data

## 4. Historical Perspective

### Timeline of NLP Evolution

**1950s-1960s: Rule-Based Systems**
- Machine Translation: Georgetown-IBM experiment (1954)
- ELIZA chatbot (1966)
- Hand-crafted rules and dictionaries

**1970s-1980s: Statistical Revolution**
- Hidden Markov Models for POS tagging
- Statistical machine translation
- Introduction of corpora and evaluation metrics

**1990s-2000s: Machine Learning Era**
- Support Vector Machines, Maximum Entropy models
- Feature engineering becomes crucial
- Supervised learning dominates

**2010s: Deep Learning Revolution**
- Word embeddings (Word2Vec, GloVe)
- Recurrent Neural Networks (RNNs, LSTMs)
- Attention mechanisms

**2017-Present: Transformer Era**
- BERT, GPT, T5, and beyond
- Pre-training + fine-tuning paradigm
- Large Language Models (LLMs)

### Key Paradigm Shifts

1. **Rules → Statistics → Machine Learning → Deep Learning**
2. **Task-specific → Multi-task → Foundation models**
3. **Symbolic → Distributed representations**
4. **Feature engineering → Representation learning**
